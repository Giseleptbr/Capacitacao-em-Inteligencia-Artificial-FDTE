## Imports e Configuração

````python
# @title
# Imports e configurações
import otter
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report
)

import nltk
from nltk.corpus import stopwords

nltk.download('stopwords', quiet=True)

# Configurações
import warnings
warnings.filterwarnings('ignore')

# Stopwords em português
STOPWORDS_PT = stopwords.words('portuguese')

# Seed para reprodutibilidade
random.seed(42)
np.random.seed(42)

# Inicializar grader
grader = otter.Notebook()

print("Setup completo!")
````

## Carregamento dos Dados

````python
# Carregar dataset B2W-Reviews (NÃO MODIFIQUE)
URL_DATASET = "https://raw.githubusercontent.com/americanas-tech/b2w-reviews01/main/B2W-Reviews01.csv"

df = pd.read_csv(URL_DATASET, nrows=6000)
df = df.dropna(subset=['review_text'])

# Criar labels de sentimento (NÃO MODIFIQUE)
df_sentimento = df[df['overall_rating'] != 3].copy()
df_sentimento['label'] = (df_sentimento['overall_rating'] >= 4).astype(int)

print(f"Dataset carregado: {len(df)} reviews")
print(f"Dataset para classificação: {len(df_sentimento)} reviews")
````

````python
# Preparar dados para exercícios (NÃO MODIFIQUE)

# Features e labels
X_TEXTOS = df_sentimento['review_text'].values
Y_LABELS = df_sentimento['label'].values

# Split treino/teste com seed fixo
X_TREINO, X_TESTE, Y_TREINO, Y_TESTE = train_test_split(
    X_TEXTOS, Y_LABELS,
    test_size=0.2,
    random_state=42,
    stratify=Y_LABELS
)

print(f"Treino: {len(X_TREINO)} | Teste: {len(X_TESTE)}")
print(f"\nDistribuição no treino:")
print(f"  Positivos (1): {(Y_TREINO == 1).sum()}")
print(f"  Negativos (0): {(Y_TREINO == 0).sum()}")
````

## Questão 5: Função de Métricas e Comparação de Modelos

Implemente uma função que calcula todas as métricas de uma só vez e use-a para comparar os modelos.

**Você deve criar:**
1. `calcular_metricas_classificacao(y_real, y_pred)`: Função que retorna dicionário com métricas
2. `metricas_comparacao_q5`: Dicionário com métricas de todos os 3 modelos

**Formato do retorno da função:**
```python
{'acuracia': float, 'precisao': float, 'recall': float, 'f1': float}
```

**Formato de `metricas_comparacao_q5`:**
```python
{
    'NaiveBayes': {'acuracia': ..., 'precisao': ..., 'recall': ..., 'f1': ...},
    'LogisticRegression': {...},
    'SVM': {...}
}
```

````python
def calcular_metricas_classificacao(y_real, y_pred):
    """
    Calcula métricas de classificação binária.

    Parâmetros:
        y_real: array com valores reais
        y_pred: array com valores preditos

    Retorna:
        dict: dicionário com acuracia, precisao, recall e f1
    """
# === SEU CÓDIGO AQUI ===
    return {
        'acuracia': accuracy_score(y_real, y_pred),
        'precisao': precision_score(y_real, y_pred),
        'recall': recall_score(y_real, y_pred),
        'f1': f1_score(y_real, y_pred)
    }

metricas_comparacao_q5 = {
    'NaiveBayes': calcular_metricas_classificacao(Y_TESTE, predicoes_nb_q2),
    'LogisticRegression': calcular_metricas_classificacao(Y_TESTE, predicoes_logreg_q3),
    'SVM': calcular_metricas_classificacao(Y_TESTE, predicoes_svm_q4)
}
````

## Questão 9: Tratamento de Negação

Um dos principais desafios em análise de sentimentos é lidar com negações. A frase "não é bom" deveria ser negativa, mas modelos bag-of-words tratam "não" e "bom" como palavras independentes.

**Solução:** Adicionar prefixo `NEG_` às palavras que seguem uma negação até encontrar pontuação.

**Você deve criar:**

1. `tratar_negacao(texto)`: Função que adiciona prefixo `NEG_` após palavras de negação
2. `X_treino_negacao_q9`: Lista com textos de treino tratados
3. `X_teste_negacao_q9`: Lista com textos de teste tratados
4. `f1_sem_negacao_q9`: F1-Score do modelo SEM tratamento (use `predicoes_logreg_q3`)
5. `f1_com_negacao_q9`: F1-Score do modelo COM tratamento de negação

**Especificação da função `tratar_negacao`:**
- Palavras de negação: `['não', 'nunca', 'nem', 'jamais', 'nenhum']`
- Adicionar `NEG_` a todas as palavras após uma negação
- Parar de adicionar `NEG_` ao encontrar pontuação: `.!?,;`
- Converter texto para minúsculas

**Exemplo:**
```
"Não gostei do produto" → "não NEG_gostei NEG_do NEG_produto"
"Bom, mas não recomendo." → "bom, mas não NEG_recomendo."
```

**Para calcular `f1_com_negacao_q9`:**
- Crie um novo TfidfVectorizer e LogisticRegression
- Treine nos dados com negação tratada
- Use os mesmos parâmetros da Q1 e Q3

````python
def tratar_negacao(texto):
    """Adiciona prefixo NEG_ após palavras de negação, até pontuação final."""
    # === SEU CÓDIGO AQUI ===

    negacoes = ['não', 'nunca', 'nem', 'jamais', 'nenhum']
    pontuacoes = '.!?,;'

    palavras = texto.lower().split()
    resultado = []
    negando = False

    for palavra in palavras:
        limpa = palavra.strip(pontuacoes)

        if limpa in negacoes:
            negando = True
            resultado.append(limpa)
            continue

        if negando:
            resultado.append(f"NEG_{palavra}")
            if any(p in palavra for p in pontuacoes):
                negando = False
        else:
            resultado.append(palavra)

    return " ".join(resultado)

X_treino_negacao_q9 = [tratar_negacao(t) for t in X_TREINO]
X_teste_negacao_q9 = [tratar_negacao(t) for t in X_TESTE]

f1_sem_negacao_q9 = f1_score(Y_TESTE, predicoes_logreg_q3)

vetor_neg = TfidfVectorizer(
    max_features=4000,
    ngram_range=(1, 2),
    min_df=3,
    stop_words=STOPWORDS_PT
)

Xtr_neg = vetor_neg.fit_transform(X_treino_negacao_q9)
Xte_neg = vetor_neg.transform(X_teste_negacao_q9)

modelo_neg = LogisticRegression(C=0.5, max_iter=1000, random_state=42)
modelo_neg.fit(Xtr_neg, Y_TREINO)

pred_neg = modelo_neg.predict(Xte_neg)
f1_com_negacao_q9 = f1_score(Y_TESTE, pred_neg)
````
