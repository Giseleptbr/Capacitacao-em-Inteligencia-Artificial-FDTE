## Questão 1: Função de Extração de Entidades

Implemente a função `extrair_entidades_spacy` que extrai entidades nomeadas de um texto usando spaCy.

**Requisitos da função:**
1. Receber um texto (string) como entrada
2. Processar o texto com o modelo `nlp` carregado
3. Retornar um dicionário com:
   - `'entidades'`: Lista de dicionários, cada um com `'texto'` e `'tipo'`
   - `'por_tipo'`: Dicionário agrupando entidades por tipo
   - `'total'`: Número total de entidades encontradas

**Assinatura:** `extrair_entidades_spacy(texto: str) -> dict`

````python
def extrair_entidades_spacy(texto):
    """
    Extrai entidades nomeadas de um texto usando spaCy.

    Parâmetros:
        texto: String com o texto a ser processado

    Retorna:
        dict: Dicionário com 'entidades', 'por_tipo' e 'total'
    """
    # === SEU CÓDIGO AQUI ===
    doc = nlp(texto)

    entidades = []
    por_tipo = defaultdict(list)

    for ent in doc.ents:
        entidades.append({"texto": ent.text, "tipo": ent.label_})
        por_tipo[ent.label_].append(ent.text)

    return {
        "entidades": entidades,
        "por_tipo": dict(por_tipo),
        "total": len(entidades)
    }
````

## Questão 3: Matcher para Legislação

Crie um `Matcher` do spaCy para identificar referências a legislação brasileira.

**Você deve criar:**
1. `matcher_legislacao`: Um objeto `Matcher` com padrão para capturar:
   - "Lei nº NÚMERO"
   - "Decreto nº NÚMERO"
   - "Resolução nº NÚMERO"
   - "Lei Complementar nº NÚMERO"
   
2. `resultados_legislacao`: Lista com os textos de legislação encontrados em `TEXTOS_JURIDICOS[0]`

**Dicas:**
- Use `{"LOWER": {"IN": [...]}}` para aceitar múltiplas palavras
- Crie mais de um padrão para facilitar matches complexos. Múltiplos padrões podem ser adicionados ao matcher com `.add("PALAVRA-CHAVE", [padrao1, padrao2, padrao3])`

````python
# === SEU CÓDIGO AQUI ===

matcher_legislacao = Matcher(nlp.vocab)

padrao_numero = [
    {"LIKE_NUM": True},
    {"TEXT": "/", "OP": "?"},
    {"LIKE_NUM": True, "OP": "?"}
]

padrao_lei = [
    {"LOWER": "lei"},
    {"LOWER": {"IN": ["nº", "no", "numero", "n"]}, "OP": "?"},
    *padrao_numero
]

padrao_decreto = [
    {"LOWER": "decreto"},
    {"LOWER": {"IN": ["nº", "no", "numero", "n"]}, "OP": "?"},
    *padrao_numero
]

padrao_resolucao = [
    {"LOWER": "resolução"},
    {"LOWER": {"IN": ["nº", "no", "numero", "n"]}, "OP": "?"},
    *padrao_numero
]

padrao_lei_complementar = [
    {"LOWER": "lei"},
    {"LOWER": "complementar"},
    {"LOWER": {"IN": ["nº", "no", "numero", "n"]}, "OP": "?"},
    *padrao_numero
]

matcher_legislacao.add(
    "LEGISLACAO",
    [padrao_lei, padrao_decreto, padrao_resolucao, padrao_lei_complementar]
)

doc0 = nlp(TEXTOS_JURIDICOS[0])

matches = matcher_legislacao(doc0)

resultados_legislacao = []
for match_id, start, end in matches:
    resultados_legislacao.append(doc0[start:end].text)

````

## Questão 4: PhraseMatcher para Tribunais

Crie um `PhraseMatcher` para encontrar menções a tribunais usando a lista `LISTA_TRIBUNAIS`.

**Você deve criar:**
1. `phrase_matcher_tribunais`: Um `PhraseMatcher` configurado com `attr="LOWER"` contendo todos os tribunais da lista

2. `buscar_tribunais(texto)`: Função que recebe um texto e retorna lista de tribunais encontrados

**Dica:** Use `nlp.make_doc(termo)` para criar os padrões do PhraseMatcher

````python
# === SEU CÓDIGO AQUI ===

phrase_matcher_tribunais = PhraseMatcher(nlp.vocab, attr="LOWER")

patterns = [nlp.make_doc(t) for t in LISTA_TRIBUNAIS]
phrase_matcher_tribunais.add("TRIBUNAIS", patterns)


def buscar_tribunais(texto):
    doc = nlp(texto)
    matches = phrase_matcher_tribunais(doc)

    resultados = []
    for match_id, start, end in matches:
        resultados.append(doc[start:end].text)

    return list(set(resultados))
````

## Questão 5: Converter Anotação BIO para Entidades

Implemente a função `converter_bio_para_entidades` que converte tokens anotados no formato BIO para entidades completas.

**Formato BIO:**
- `B-TIPO`: Início de entidade do tipo TIPO
- `I-TIPO`: Continuação de entidade do tipo TIPO  
- `O`: Token fora de entidade

**Requisitos:**
1. Receber lista de tokens e lista de tags BIO
2. Juntar tokens consecutivos B- e I- do mesmo tipo
3. Retornar lista de tuplas `(texto_entidade, tipo)`

**Exemplo:**
```python
tokens = ["João", "Silva", "mora", "em", "São", "Paulo"]
tags = ["B-PER", "I-PER", "O", "O", "B-LOC", "I-LOC"]
# Resultado: [("João Silva", "PER"), ("São Paulo", "LOC")]
```

````python
def converter_bio_para_entidades(tokens, tags):
    """
    Converte anotação BIO para lista de entidades.

    Parâmetros:
        tokens: Lista de tokens (strings)
        tags: Lista de tags BIO (strings)

    Retorna:
        list: Lista de tuplas (texto_entidade, tipo)
    """
    # === SEU CÓDIGO AQUI ===
    entidades = []

    entidade_atual = []
    tipo_atual = None

    for token, tag in zip(tokens, tags):
        if tag.startswith("B-"):
            if entidade_atual:
                entidades.append((" ".join(entidade_atual), tipo_atual))

            entidade_atual = [token]
            tipo_atual = tag.split("-")[1]

        elif tag.startswith("I-") and tipo_atual == tag.split("-")[1]:
            entidade_atual.append(token)

        else:
            if entidade_atual:
                entidades.append((" ".join(entidade_atual), tipo_atual))
                entidade_atual = []
                tipo_atual = None

    if entidade_atual:
        entidades.append((" ".join(entidade_atual), tipo_atual))

    return entidades
````

   - `'tribunais'`: Lista de tribunais encontrados (usando PhraseMatcher)

**Use os matchers criados nas questões anteriores como referência!**

````python
class ExtratorInformacaoJuridico:
    """
    Sistema de Extração de Informação para textos jurídicos.

    Combina NER, Matcher e PhraseMatcher para extrair:
    - Entidades nomeadas (pessoas, organizações, locais)
    - Referências a legislação
    - Menções a tribunais
    """
    # === SEU CÓDIGO AQUI ===

    def __init__(self):
        """
        Inicializa o extrator com modelo spaCy e matchers.
        """
        self.nlp = nlp
        self.matcher_legislacao = matcher_legislacao
        self.phrase_matcher_tribunais = phrase_matcher_tribunais

    def processar(self, texto):
        """
        Processa um texto e extrai informações.

        Parâmetros:
            texto: String com o texto jurídico

        Retorna:
            dict: Dicionário com 'entidades', 'legislacao' e 'tribunais'
        """


        doc = self.nlp(texto)

        # NER
        entidades = []
        for ent in doc.ents:
            entidades.append({
                "texto": ent.text,
                "tipo": ent.label_
            })

        # Legislação
        matches_leg = self.matcher_legislacao(doc)
        legislacao = [doc[start:end].text for _, start, end in matches_leg]

        # Tribunais
        matches_tri = self.phrase_matcher_tribunais(doc)
        tribunais = [doc[start:end].text for _, start, end in matches_tri]

        return {
            "entidades": entidades,
            "legislacao": list(set(legislacao)),
            "tribunais": list(set(tribunais))
        }
````
