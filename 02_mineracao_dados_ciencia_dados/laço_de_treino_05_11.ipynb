```python
import pandas as pd
from river.tree import HoeffdingTreeRegressor
from river import metrics, stream
from ucimlrepo import fetch_ucirepo
```

```python
abalone = fetch_ucirepo(id=1)
X = abalone.data.features.copy()
y = abalone.data.targets["Rings"]
```

```python
# Converte a variГЎvel categГіrica 'Sex' em numГ©rica (sem warnings)
X.loc[:, "Sex"] = X["Sex"].map({'M': 0, 'F': 1, 'I': 2}).astype(float)
```

```python
model = HoeffdingTreeRegressor()
rmse = metrics.RMSE()

for x, yi in stream.iter_pandas(X, y):
    y_pred = model.predict_one(x)
    rmse.update(yi, y_pred)
    model.learn_one(x, yi)
```
