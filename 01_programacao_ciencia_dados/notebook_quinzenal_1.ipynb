## Contexto

Cada dicionário tem as chaves: `'nome'`, `'idade'`, e `'cidade'`.

## Tarefa
Crie uma função que analisa este dataset e retorna um dicionário com:

1. `'total'`: número total de pessoas
2. `'idade_media'`: média das idades
3. `'idade_min'`: menor idade
4. `'idade_max'`: maior idade
5. `'cidades'`: dicionário com contagem de pessoas por cidade
6. `'maiores_de_idade'`: lista com nomes das pessoas com idade >= 18

## Exemplo
```python
dados = [
    {'nome': 'Ana', 'idade': 25, 'cidade': 'SP'},
    {'nome': 'Bruno', 'idade': 17, 'cidade': 'RJ'},
    {'nome': 'Carlos', 'idade': 30, 'cidade': 'SP'}

resultado = analisar_dataset(dados)

#     'cidades': {'SP': 2, 'RJ': 1},
#     'maiores_de_idade': ['Ana', 'Carlos']
# }
```

## Requisitos
- Use compreensão de listas/dicionários quando apropriado
- Use `Counter` para contar cidades
- Arredonde a idade média para 2 casas decimais

## Pontos de Avaliação
- Corretude dos cálculos (10 pts)
- Uso de estruturas adequadas (3 pts)
- Código limpo (2 pts)

**Tempo estimado:** 10 minutos

````python
from collections import Counter

def analisar_dataset(dados):
    """
    Analisa dataset de pessoas.

    Args:
        dados (list): Lista de dicionários com chaves 'nome', 'idade', 'cidade'

    Returns:
        dict: Dicionário com análises conforme especificação
    """
    # === SEU CÓDIGO AQUI ===
    total = len(dados)
    idades = [p['idade'] for p in dados]
    idade_media = round(sum(idades) / total, 2)
    idade_min = min(idades)
    idade_max = max(idades)
    cidades = dict(Counter([p['cidade'] for p in dados]))
    maiores_de_idade = [p['nome'] for p in dados if p['idade'] >= 18]

    return {
        'total': total,
        'idade_media': idade_media,
        'idade_min': idade_min,
        'idade_max': idade_max,
        'cidades': cidades,
        'maiores_de_idade': maiores_de_idade
    }

    # === FIM DO SEU CÓDIGO ===
````

# Questão 2: Processamento de Outliers (20 pontos)

## Contexto
Em análise de dados, outliers (valores atípicos) podem distorcer análises estatísticas.
Uma técnica comum é identificá-los usando o método Z-score e substituí-los.

## Conceito: Z-Score
O Z-score mede quantos desvios padrões um valor está da média:
```
z = (x - μ) / σ
```

Se |z| > threshold, o valor é considerado outlier.

## Tarefa
Implemente uma função que:
1. Calcula o Z-score de cada valor
2. Identifica outliers (|z-score| > threshold)
3. Substitui outliers pela **mediana** dos valores não-outliers
4. Retorna o array processado

## Exemplo
```python
dados = np.array([10, 12, 11, 13, 100, 14, 12, 11])
# 100 é outlier óbvio

resultado = processar_outliers(dados, threshold=2.0)
# 100 será substituído pela mediana dos outros valores (~12)
```

## Requisitos
- Use operações vetorizadas NumPy
- **NÃO use loops**
- Use `np.abs()`, `np.where()`, `np.median()`

## Pontos de Avaliação
- Cálculo correto do Z-score (8 pts)
- Identificação de outliers (5 pts)
- Substituição correta (5 pts)
- Uso de vetorização (2 pts)

**Tempo estimado:** 12 minutos

````python
def processar_outliers(dados, threshold=2.0):
    """
    Remove outliers substituindo pela mediana.

    Args:
        dados (np.ndarray): Array 1D de valores
        threshold (float): Threshold do Z-score para outlier

    Returns:
        np.ndarray: Array com outliers substituídos
    """
    # === SEU CÓDIGO AQUI ===

    media = np.mean(dados)
    desvio = np.std(dados)
    z_scores = (dados - media) / desvio
    nao_outliers = dados[np.abs(z_scores) <= threshold]
    mediana = np.median(nao_outliers)
    return np.where(np.abs(z_scores) > threshold, mediana, dados)


    # === FIM DO SEU CÓDIGO ===
````

# Questão 3: Regressão Linear - Equação Normal (25 pontos)

## Contexto
A regressão linear é fundamental em Machine Learning.
Você vai implementá-la do zero usando álgebra linear.

## Teoria: Equação Normal
Para encontrar os parâmetros θ que minimizam o erro:

```
θ = (X^T X)^-1 X^T y
```

Onde:
- X é a matriz de features (com coluna de 1s para intercepto)
- y é o vetor de targets
- θ contém [intercepto, coef1, coef2, ...]

## Tarefa
Implemente regressão linear que:
1. Adiciona coluna de 1s ao início de X (para intercepto)
2. Calcula θ usando a equação normal
3. Separa intercepto e coeficientes
4. Calcula R² (coeficiente de determinação)
5. Retorna dicionário com todos os resultados

## Fórmula do R²
```
R² = 1 - (SS_res / SS_tot)

onde:
SS_res = Σ(y - y_pred)²  # Soma dos quadrados dos resíduos
SS_tot = Σ(y - y_mean)²  # Soma total dos quadrados
```

## Exemplo
```python
# Dados: y = 2*x1 + 3*x2 + 5 + ruído
X = np.array([[1, 2], [2, 3], [3, 4]])
y = np.array([12, 17, 22])  # aproximadamente

resultado = regressao_linear_normal_equation(X, y)
# {
#     'intercepto': ~5,
#     'coeficientes': [~2, ~3],
#     'r_squared': ~0.99,
#     'theta': [~5, ~2, ~3]
# }
```

## Requisitos
- Use `np.linalg.inv()` e operador `@`
- Use `np.c_[]` ou `np.column_stack()` para adicionar coluna
- **NÃO use bibliotecas de ML** (sklearn, statsmodels, etc)

## Pontos de Avaliação
- Adição correta da coluna de 1s (4 pts)
- Cálculo correto de θ (10 pts)
- Separação correta de intercepto/coef (3 pts)
- Cálculo correto de R² (6 pts)
- Estrutura do retorno (2 pts)

**Tempo estimado:** 15 minutos

````python
def regressao_linear_normal_equation(X, y):
    """
    Implementa regressão linear usando equação normal.

    Args:
        X (np.ndarray): Matriz de features (n_samples, n_features)
        y (np.ndarray): Vetor de targets (n_samples,)

    Returns:
        dict: {
            'intercepto': float,
            'coeficientes': np.ndarray,
            'r_squared': float,
            'theta': np.ndarray  # [intercepto, coef1, coef2, ...]
        }
    """
    # === SEU CÓDIGO AQUI ===

    X_b = np.c_[np.ones((X.shape[0], 1)), X]
    theta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
    intercepto = theta[0]
    coeficientes = theta[1:]
    y_pred = X_b @ theta
    ss_res = np.sum((y - y_pred) ** 2)
    ss_tot = np.sum((y - np.mean(y)) ** 2)
    r_squared = 1 - ss_res / ss_tot
    return {
        'intercepto': intercepto,
        'coeficientes': coeficientes,
        'r_squared': r_squared,
        'theta': theta
    }

    # === FIM DO SEU CÓDIGO ===
````

````python
# Teste sua solução
print("=== TESTE Q3 ===")

# Gerar dados sintéticos: y = 3*x1 + 5*x2 + 2 + ruído
np.random.seed(42)
n_samples = 100
X_teste = np.random.randn(n_samples, 2) * 10
y_verdadeiro = 3 * X_teste[:, 0] + 5 * X_teste[:, 1] + 2
y_teste = y_verdadeiro + np.random.randn(n_samples) * 2  # ruído

print("Dados gerados:")
print(f"  Relação verdadeira: y = 3*x1 + 5*x2 + 2")
print(f"  Amostras: {n_samples}")
print(f"  Features: {X_teste.shape[1]}")

# Treinar modelo
resultado = regressao_linear_normal_equation(X_teste, y_teste)

print("\n=== RESULTADOS ===")
print(f"Intercepto estimado: {resultado['intercepto']:.4f} (esperado: ~2.0)")
print(f"Coeficientes estimados:")
for i, coef in enumerate(resultado['coeficientes']):
    esperado = [3.0, 5.0][i]
    print(f"  x{i+1}: {coef:.4f} (esperado: ~{esperado})")
print(f"\nR² Score: {resultado['r_squared']:.4f}")
print(f"Theta completo: {resultado['theta']}")

# Verificar qualidade
if resultado['r_squared'] > 0.95:
    print("\n✓ Excelente ajuste! R² > 0.95")
elif resultado['r_squared'] > 0.90:
    print("\n✓ Bom ajuste! R² > 0.90")
else:
    print("\n⚠ Ajuste pode melhorar. Verifique sua implementação.")

# Executar testes públicos
grader.check("q3")
````

# Questão 4: Preparação de Dados para ML (20 pontos)

## Contexto
Antes de treinar modelos, dados precisam ser preparados:
normalização, divisão train/test, e remoção de features correlacionadas.

## Tarefa
Implemente uma função que prepara dados para Machine Learning:

1. **Normalização Z-score** de todas as colunas
2. **Remoção de features correlacionadas:**
   - Calcule matriz de correlação
   - Se duas features têm |correlação| > 0.95, remova uma delas
3. **Divisão train/test:**
   - Split baseado na proporção `train_size`
   - **NÃO embaralhe** os dados (manter ordem)
4. **Retorne:** tupla (X_train, X_test)

## Exemplo
```python
# Dados com 100 amostras, 5 features
dados = np.random.randn(100, 5)

X_train, X_test = preparar_dados_ml(dados, train_size=0.8)
# X_train: (80, n_features_após_remoção)
# X_test: (20, n_features_após_remoção)
```

## Requisitos Técnicos
- Use `np.corrcoef()` para correlação
- Use `np.triu_indices()` para iterar na matriz triangular superior
- Normalização: `(X - mean) / std` por coluna
- Split sem shuffle: fatiamento simples

## Pontos de Avaliação
- Normalização correta (6 pts)
- Identificação de features correlacionadas (6 pts)
- Remoção correta de features (4 pts)
- Split train/test correto (4 pts)

**Tempo estimado:** 10 minutos

````python
def preparar_dados_ml(dados, train_size=0.8):
    """
    Prepara dados para Machine Learning.

    Args:
        dados (np.ndarray): Matriz de features (n_samples, n_features)
        train_size (float): Proporção para treino (0 < train_size < 1)

    Returns:
        tuple: (X_train, X_test) ambos normalizados e sem features correlacionadas
    """
    # === SEU CÓDIGO AQUI ===

    # Normalização Z-score
    mean = np.mean(dados, axis=0)
    std = np.std(dados, axis=0)
    X_norm = (dados - mean) / std

    # Remoção de features correlacionadas
    corr = np.corrcoef(X_norm, rowvar=False)
    to_remove = set()
    n_features = corr.shape[0]
    for i in range(n_features):
        for j in range(i + 1, n_features):
            if abs(corr[i, j]) > 0.95:
                to_remove.add(j)
    mask = [i for i in range(n_features) if i not in to_remove]
    X_reduzido = X_norm[:, mask]

    # Split train/test
    n_train = int(len(X_reduzido) * train_size)
    X_train = X_reduzido[:n_train]
    X_test = X_reduzido[n_train:]
    return X_train, X_test

    # === FIM DO SEU CÓDIGO ===
````

````python
# Teste sua solução
print("=== TESTE Q4 ===")

# Gerar dados com features correlacionadas
np.random.seed(42)
n_samples = 200

# Feature 1: independente
f1 = np.random.randn(n_samples)

# Feature 2: independente
f2 = np.random.randn(n_samples) * 2

# Feature 3: altamente correlacionada com f1 (quase duplicata)
f3 = f1 + np.random.randn(n_samples) * 0.01

# Feature 4: independente
f4 = np.random.randn(n_samples) * 0.5

# Feature 5: altamente correlacionada com f2
f5 = f2 * 1.01 + np.random.randn(n_samples) * 0.01

dados_teste = np.column_stack([f1, f2, f3, f4, f5])

print(f"Dados originais:")
print(f"  Shape: {dados_teste.shape}")
print(f"  Features: {dados_teste.shape[1]}")

# Mostrar correlações
corr = np.corrcoef(dados_teste, rowvar=False)
print(f"\n  Correlações altas detectadas:")
for i in range(5):
    for j in range(i+1, 5):
        if abs(corr[i, j]) > 0.95:
            print(f"    f{i+1} <-> f{j+1}: {corr[i,j]:.4f}")

# Preparar dados
X_train, X_test = preparar_dados_ml(dados_teste, train_size=0.75)

print(f"\n=== RESULTADOS ===")
print(f"Treino:")
print(f"  Shape: {X_train.shape}")
print(f"  Média por feature: {X_train.mean(axis=0)}")
print(f"  Desvio padrão: {X_train.std(axis=0)}")

print(f"\nTeste:")
print(f"  Shape: {X_test.shape}")

print(f"\nVerificações:")
print(f"  ✓ Features removidas: {5 - X_train.shape[1]}")
print(f"  ✓ Total de amostras preservado: {len(X_train) + len(X_test) == n_samples}")
print(f"  ✓ Proporção train/test: {len(X_train)/n_samples:.2f} / {len(X_test)/n_samples:.2f}")

# Verificar normalização
if np.allclose(X_train.mean(axis=0), 0, atol=0.1):
    print(f"  ✓ Dados normalizados (média ≈ 0)")
if np.allclose(X_train.std(axis=0), 1, atol=0.1):
    print(f"  ✓ Dados normalizados (std ≈ 1)")

# Executar testes públicos
grader.check("q4")
````

# Questão 5: Projeto Integrador (20 pontos)

## Contexto
Esta questão integra todos os conceitos das aulas 1-4.
Você vai criar um pipeline completo de análise de dados.

## Tarefa
Implemente uma função que recebe uma matriz de dados e retorna
uma análise estatística completa:

### 1. Estatísticas Descritivas (6 pts)
Para cada coluna calcule:
- média, mediana, desvio padrão
- mínimo, máximo
- percentis 25, 50, 75

### 2. Matriz de Correlação (5 pts)
- Calcule correlação entre todas as colunas
- Use `np.corrcoef()`

### 3. Detecção de Outliers (5 pts)
Para cada coluna:
- Identifique outliers usando IQR method:
- `outlier se: valor < Q1 - 1.5*IQR OU valor > Q3 + 1.5*IQR`
- Conte quantos outliers em cada coluna

### 4. Análise de Componentes (4 pts)
- Calcule autovalores da matriz de correlação
- Ordene do maior para o menor
- Calcule variância explicada por cada componente

## Formato de Retorno
```python
{
    'estatisticas': {
        'mean': array([...]),
        'median': array([...]),
        'std': array([...]),
        'min': array([...]),
        'max': array([...]),
        'q25': array([...]),
        'q50': array([...]),
        'q75': array([...])
    },
    'matriz_correlacao': array([[...]]),
    'outliers_por_coluna': array([...]),  # contagem
    'autovalores': array([...]),
    'variancia_explicada': array([...])  # em porcentagem
}
```

## Requisitos
- Use apenas NumPy
- Operações vetorizadas quando possível
- Código bem organizado

## Dicas
- `np.percentile()` para quartis
- `np.corrcoef()` para correlação
- `np.linalg.eig()` para autovalores e autovetores
- IQR = Q3 - Q1
- Fórmula para variância explicada: array de autovalores / soma dos autovalores

**Tempo estimado:** 13 minutos

````python
def analise_completa(matriz):
    """
    Realiza análise estatística completa de uma matriz.

    Args:
        matriz (np.ndarray): Matriz de dados (n_samples, n_features)

    Returns:
        dict: Dicionário com todas as análises conforme especificação
    """
    # === SEU CÓDIGO AQUI ===

    mean = np.mean(matriz, axis=0)
    median = np.median(matriz, axis=0)
    std = np.std(matriz, axis=0)
    min_v = np.min(matriz, axis=0)
    max_v = np.max(matriz, axis=0)
    q25 = np.percentile(matriz, 25, axis=0)
    q50 = np.percentile(matriz, 50, axis=0)
    q75 = np.percentile(matriz, 75, axis=0)

    estatisticas = {
        'mean': mean,
        'median': median,
        'std': std,
        'min': min_v,
        'max': max_v,
        'q25': q25,
        'q50': q50,
        'q75': q75
    }

    matriz_corr = np.corrcoef(matriz, rowvar=False)

    # Outliers (IQR method)
    iqr = q75 - q25
    lower = q25 - 1.5 * iqr
    upper = q75 + 1.5 * iqr
    outliers_por_coluna = np.sum((matriz < lower) | (matriz > upper), axis=0)

    # Análise de componentes (autovalores e variância explicada)
    autovalores, _ = np.linalg.eig(matriz_corr)
    autovalores = np.sort(autovalores)[::-1]
    variancia_explicada = (autovalores / np.sum(autovalores)) * 100

    return {
        'estatisticas': estatisticas,
        'matriz_correlacao': matriz_corr,
        'outliers_por_coluna': outliers_por_coluna,
        'autovalores': autovalores,
        'variancia_explicada': variancia_explicada
    }

    # === FIM DO SEU CÓDIGO ===
````

````python
# Teste sua solução
print("=== TESTE Q5 ===")

# Gerar dados de teste
np.random.seed(42)
n_samples = 1000
n_features = 5

# Criar dados com diferentes distribuições
dados_teste = np.column_stack([
    np.random.normal(100, 15, n_samples),      # Feature 1: normal
    np.random.exponential(10, n_samples),       # Feature 2: exponencial
    np.random.uniform(0, 50, n_samples),        # Feature 3: uniforme
    np.random.normal(50, 10, n_samples),        # Feature 4: normal
    np.random.normal(25, 5, n_samples)          # Feature 5: normal
])

# Adicionar alguns outliers artificiais
dados_teste[0, 0] = 500  # Outlier na feature 1
dados_teste[1, 2] = -100  # Outlier na feature 3

print(f"Dados de teste: {dados_teste.shape}")

# Executar análise
resultado = analise_completa(dados_teste)

# Exibir resultados
print("\n=== ESTATÍSTICAS DESCRITIVAS ===")
for i in range(n_features):
    print(f"\nFeature {i+1}:")
    print(f"  Média: {resultado['estatisticas']['mean'][i]:.2f}")
    print(f"  Mediana: {resultado['estatisticas']['median'][i]:.2f}")
    print(f"  Desvio: {resultado['estatisticas']['std'][i]:.2f}")
    print(f"  Range: [{resultado['estatisticas']['min'][i]:.2f}, {resultado['estatisticas']['max'][i]:.2f}]")
    print(f"  Quartis: Q1={resultado['estatisticas']['q25'][i]:.2f}, "
          f"Q2={resultado['estatisticas']['q50'][i]:.2f}, "
          f"Q3={resultado['estatisticas']['q75'][i]:.2f}")

print("\n=== MATRIZ DE CORRELAÇÃO ===")
print(resultado['matriz_correlacao'].round(3))

print("\n=== OUTLIERS POR COLUNA ===")
for i, count in enumerate(resultado['outliers_por_coluna']):
    print(f"  Feature {i+1}: {count} outliers ({count/n_samples*100:.1f}%)")

print("\n=== ANÁLISE DE COMPONENTES ===")
print(f"Autovalores: {resultado['autovalores'].round(4)}")
print(f"\nVariância explicada por componente:")
for i, var in enumerate(resultado['variancia_explicada']):
    print(f"  PC{i+1}: {var:.2f}%")

cum_var = np.cumsum(resultado['variancia_explicada'])
print(f"\nVariância acumulada:")
for i, var in enumerate(cum_var):
    print(f"  Primeiros {i+1} PCs: {var:.2f}%")

# Executar testes públicos
grader.check("q5")
````
