# Questão 1: Preparação e Limpeza de Dados (8 pontos)

## Objetivo
Preparar os dados para análise, tratando problemas de qualidade.

## Tarefa
Implemente a função `preparar_dados()` que:

1. **Combina os DataFrames** de vendas e clientes (merge)
2. **Trata valores missing:**
   - Remove linhas onde `valor` é NaN
   - Documenta quantas linhas foram removidas
3. **Cria novas colunas derivadas:**
   - `valor_total` = valor × quantidade
   - `mes` = mês extraído da data
   - `faixa_etaria` = categorização da idade:
     - 'Jovem': < 30
     - 'Adulto': 30-50
     - 'Senior': > 50
4. **Retorna dicionário com:**
   - `df_completo`: DataFrame limpo e enriquecido
   - `relatorio_limpeza`: dict com estatísticas da limpeza

## Exemplo do Relatório
```python
{
    'linhas_originais': 500,
    'linhas_removidas': 25,
    'linhas_finais': 475,
    'percentual_removido': 5.0
}
```

## Dicas
- Use `pd.merge()` para combinar os DataFrames
- Use `.dropna()` para remover missing
- Use `.apply()` ou condições para criar faixa_etaria
- Use `.dt.month` para extrair mês da data

**Tempo estimado:** 15 minutos

````python
def preparar_dados(df_vendas, df_clientes):
    """
    Prepara e limpa dados para análise.

    Args:
        df_vendas (pd.DataFrame): DataFrame de vendas
        df_clientes (pd.DataFrame): DataFrame de clientes

    Returns:
        dict: Dicionário com df_completo e relatorio_limpeza
    """
    # === SEU CÓDIGO AQUI ===

    linhas_originais = len(df_vendas)

    # Merge
    df_completo = pd.merge(df_vendas, df_clientes, on='cliente_id', how='left')

    # Remover missing em 'valor'
    df_limpo = df_completo.dropna(subset=['valor']).copy()
    linhas_removidas = linhas_originais - len(df_limpo)

    # Criar novas colunas
    df_limpo['valor_total'] = df_limpo['valor'] * df_limpo['quantidade']
    df_limpo['mes'] = df_limpo['data'].dt.month

    # Faixa etária
    def faixa_etaria(idade):
        if idade < 30:
            return 'Jovem'
        elif idade <= 50:
            return 'Adulto'
        else:
            return 'Senior'

    df_limpo['faixa_etaria'] = df_limpo['idade'].apply(faixa_etaria)

    # Relatório
    relatorio = {
        'linhas_originais': linhas_originais,
        'linhas_removidas': linhas_removidas,
        'linhas_finais': len(df_limpo),
        'percentual_removido': (linhas_removidas / linhas_originais) * 100
    }

    return {'df_completo': df_limpo, 'relatorio_limpeza': relatorio}

    # === FIM DO SEU CÓDIGO ===
````

````python
# === TESTE Q1 ===

print("=" * 70)
print("QUESTÃO 1: PREPARAÇÃO E LIMPEZA DE DADOS")
print("=" * 70)

# Preparar dados
resultado_prep = preparar_dados(df_vendas, df_clientes)
df_completo = resultado_prep['df_completo']
relatorio = resultado_prep['relatorio_limpeza']

print("\n=== RELATÓRIO DE LIMPEZA ===")
for chave, valor in relatorio.items():
    if 'percentual' in chave:
        print(f"{chave}: {valor:.2f}%")
    else:
        print(f"{chave}: {valor}")

print("\n=== AMOSTRA DOS DADOS LIMPOS ===")
print(df_completo.head())

print("\n=== INFORMAÇÕES DO DATASET ===")
print(f"Shape: {df_completo.shape}")
print(f"\nColunas: {list(df_completo.columns)}")

print("\n=== VERIFICAÇÕES ===")
print(f"✓ Missing em 'valor': {df_completo['valor'].isna().sum()}")
print(f"✓ Coluna 'valor_total' criada: {'valor_total' in df_completo.columns}")
print(f"✓ Coluna 'mes' criada: {'mes' in df_completo.columns}")
print(f"✓ Coluna 'faixa_etaria' criada: {'faixa_etaria' in df_completo.columns}")

print("\n=== DISTRIBUIÇÃO POR FAIXA ETÁRIA ===")
print(df_completo['faixa_etaria'].value_counts())

# Verificar testes
grader.check("q1")

print("\n" + "=" * 70)
print("FIM DA QUESTÃO 1")
print("=" * 70)
````

# Questão 2: Análise de Vendas Avançada (10 pontos)

## Objetivo
Realizar análise detalhada usando groupby, agregações e pivot tables.

## Tarefa
Implemente a função `analise_vendas_avancada()` que retorna dict com:

1. **por_regiao**: DataFrame com análise por região
   - total_vendas: soma de valor_total
   - num_transacoes: contagem de vendas
   - ticket_medio: média de valor_total
   - quantidade_media: média de quantidade

2. **por_categoria**: DataFrame com análise por categoria
   - Mesmas métricas que por_regiao

3. **pivot_regiao_categoria**: Pivot table
   - Linhas: região
   - Colunas: categoria
   - Valores: soma de valor_total

4. **top_clientes**: DataFrame com top 10 clientes por valor_total
   - Colunas: cliente_id, nome, total_gasto, num_compras

5. **metricas_gerais**: dict com:
   - receita_total
   - num_vendas_total
   - ticket_medio_geral
   - regiao_mais_lucrativa
   - categoria_mais_vendida

## Dicas
- Use `groupby().agg()` para múltiplas agregações
- Use `pivot_table()` para tabela cruzada
- Use `.nlargest()` para top clientes

**Tempo estimado:** 20 minutos

````python
def analise_vendas_avancada(df):
    """
    Análise avançada de vendas com groupby e pivot.

    Args:
        df (pd.DataFrame): DataFrame completo

    Returns:
        dict: Dicionário com todas as análises
    """
    # === SEU CÓDIGO AQUI ===

    # 1. Por região
    por_regiao = (
        df.groupby('regiao')
        .agg(total_vendas=('valor_total', 'sum'),
             num_transacoes=('venda_id', 'count'),
             ticket_medio=('valor_total', 'mean'),
             quantidade_media=('quantidade', 'mean'))
        .reset_index()
    )

    # 2. Por categoria
    por_categoria = (
        df.groupby('categoria')
        .agg(total_vendas=('valor_total', 'sum'),
             num_transacoes=('venda_id', 'count'),
             ticket_medio=('valor_total', 'mean'),
             quantidade_media=('quantidade', 'mean'))
        .reset_index()
    )

    # 3. Pivot table
    pivot_regiao_categoria = pd.pivot_table(
        df,
        index='regiao',
        columns='categoria',
        values='valor_total',
        aggfunc='sum',
        fill_value=0
    )

    # 4. Top 10 clientes
    top_clientes = (
        df.groupby(['cliente_id', 'nome'])
        .agg(total_gasto=('valor_total', 'sum'),
             num_compras=('venda_id', 'count'))
        .reset_index()
        .sort_values('total_gasto', ascending=False)
        .head(10)
    )

    # 5. Métricas gerais
    receita_total = df['valor_total'].sum()
    num_vendas_total = df.shape[0]
    ticket_medio_geral = df['valor_total'].mean()
    regiao_mais_lucrativa = por_regiao.loc[por_regiao['total_vendas'].idxmax(), 'regiao']
    categoria_mais_vendida = por_categoria.loc[por_categoria['total_vendas'].idxmax(), 'categoria']

    metricas_gerais = {
        'receita_total': receita_total,
        'num_vendas_total': num_vendas_total,
        'ticket_medio_geral': ticket_medio_geral,
        'regiao_mais_lucrativa': regiao_mais_lucrativa,
        'categoria_mais_vendida': categoria_mais_vendida
    }

    return {
        'por_regiao': por_regiao,
        'por_categoria': por_categoria,
        'pivot_regiao_categoria': pivot_regiao_categoria,
        'top_clientes': top_clientes,
        'metricas_gerais': metricas_gerais
    }

    # === FIM DO SEU CÓDIGO ===
````

````python
# === TESTE Q2 ===

print("=" * 70)
print("QUESTÃO 2: ANÁLISE DE VENDAS AVANÇADA")
print("=" * 70)

# Realizar análise
analise = analise_vendas_avancada(df_completo)

print("\n=== 1. ANÁLISE POR REGIÃO ===")
print(analise['por_regiao'].round(2))

print("\n=== 2. ANÁLISE POR CATEGORIA ===")
print(analise['por_categoria'].round(2))

print("\n=== 3. PIVOT TABLE: REGIÃO × CATEGORIA ===")
print(analise['pivot_regiao_categoria'].round(2))

print("\n=== 4. TOP 10 CLIENTES ===")
print(analise['top_clientes'].head(10).round(2))

print("\n=== 5. MÉTRICAS GERAIS ===")
metricas = analise['metricas_gerais']
print(f"Receita Total: R$ {metricas['receita_total']:,.2f}")
print(f"Número de Vendas: {metricas['num_vendas_total']}")
print(f"Ticket Médio: R$ {metricas['ticket_medio_geral']:.2f}")
print(f"Região Mais Lucrativa: {metricas['regiao_mais_lucrativa']}")
print(f"Categoria Mais Vendida: {metricas['categoria_mais_vendida']}")

# Visualizações
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Gráfico 1: Vendas por Região
analise['por_regiao'].plot(x='regiao', y='total_vendas', kind='bar', ax=axes[0], color='steelblue')
axes[0].set_title('Total de Vendas por Região', fontweight='bold')
axes[0].set_xlabel('Região')
axes[0].set_ylabel('Total de Vendas (R$)')
axes[0].tick_params(axis='x', rotation=45)

# Gráfico 2: Vendas por Categoria
analise['por_categoria'].plot(x='categoria', y='total_vendas', kind='bar', ax=axes[1], color='coral')
axes[1].set_title('Total de Vendas por Categoria', fontweight='bold')
axes[1].set_xlabel('Categoria')
axes[1].set_ylabel('Total de Vendas (R$)')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Verificar testes
grader.check("q2")

print("\n" + "=" * 70)
print("FIM DA QUESTÃO 2")
print("=" * 70)
````

# Questão 3: Análise Estatística Completa (10 pontos)

## Objetivo
Realizar análise estatística descritiva e identificar outliers.

## Tarefa
Implemente a função `analise_estatistica_completa()` que retorna dict com:

1. **estatisticas**: DataFrame com estatísticas para colunas numéricas
   - Colunas: media, mediana, desvio_padrao, cv, min, max, iqr
   - Linhas: cada variável numérica

2. **outliers**: dict onde chave é nome da coluna, valor é:
   - Series com os valores outliers
   - Use regra IQR

3. **distribuicoes**: dict com classificação da distribuição:
   - 'simetrica' ou 'assimetrica' para cada variável numérica
   - Base: |media - mediana| < 0.1 × desvio_padrao

4. **visualizacoes**: matplotlib Figure com 2×2 subplots:
   - [0,0]: Histogramas de valor_total
   - [0,1]: Boxplot de valor_total
   - [1,0]: Histograma de quantidade
   - [1,1]: Boxplot de quantidade

**Tempo estimado:** 15 minutos

````python
def analise_estatistica_completa(df):
    """
    Análise estatística completa com outliers e visualizações.

    Args:
        df (pd.DataFrame): DataFrame

    Returns:
        dict: Análise estatística completa
    """
    # === SEU CÓDIGO AQUI ===

    num_cols = df.select_dtypes(include=[np.number]).columns
    estatisticas = pd.DataFrame(index=num_cols)

    # Estatísticas básicas
    estatisticas['media'] = df[num_cols].mean()
    estatisticas['mediana'] = df[num_cols].median()
    estatisticas['desvio_padrao'] = df[num_cols].std()
    estatisticas['cv'] = estatisticas['desvio_padrao'] / estatisticas['media']
    estatisticas['min'] = df[num_cols].min()
    estatisticas['max'] = df[num_cols].max()
    estatisticas['iqr'] = df[num_cols].quantile(0.75) - df[num_cols].quantile(0.25)

    # Outliers via IQR
    outliers = {}
    for col in num_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lim_inf = Q1 - 1.5 * IQR
        lim_sup = Q3 + 1.5 * IQR
        outliers[col] = df[col][(df[col] < lim_inf) | (df[col] > lim_sup)]

    # Distribuições (simétrica x assimétrica)
    distribuicoes = {}
    for col in num_cols:
        media = estatisticas.loc[col, 'media']
        mediana = estatisticas.loc[col, 'mediana']
        desvio = estatisticas.loc[col, 'desvio_padrao']
        if desvio == 0 or pd.isna(desvio):
            distribuicoes[col] = 'simetrica'
        elif abs(media - mediana) < 0.1 * desvio:
            distribuicoes[col] = 'simetrica'
        else:
            distribuicoes[col] = 'assimetrica'

    # Verificar colunas disponíveis para plotagem
    col_valor = 'valor_total' if 'valor_total' in df.columns else num_cols[0]
    col_qtd = 'quantidade' if 'quantidade' in df.columns else num_cols[1]

    # Visualizações
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))

    sns.histplot(df[col_valor], kde=True, ax=axes[0, 0])
    axes[0, 0].set_title(f'Histograma - {col_valor}')

    sns.boxplot(x=df[col_valor], ax=axes[0, 1])
    axes[0, 1].set_title(f'Boxplot - {col_valor}')

    sns.histplot(df[col_qtd], kde=True, ax=axes[1, 0])
    axes[1, 0].set_title(f'Histograma - {col_qtd}')

    sns.boxplot(x=df[col_qtd], ax=axes[1, 1])
    axes[1, 1].set_title(f'Boxplot - {col_qtd}')

    plt.tight_layout()

    return {
        'estatisticas': estatisticas,
        'outliers': outliers,
        'distribuicoes': distribuicoes,
        'visualizacoes': fig
    }

    # === FIM DO SEU CÓDIGO ===
````

# Questão 4: Dashboard Final de EDA (12 pontos)

## Objetivo
Criar dashboard visual completo com análise de correlações e insights.

## Tarefa
Implemente a função `dashboard_final_eda()` que retorna dict com:

1. **correlacoes**: dict com:
   - `matriz`: DataFrame com matriz de correlação
   - `pares_fortes`: lista de tuplas (var1, var2, corr) onde |corr| >= 0.5
   - `heatmap`: matplotlib Figure com heatmap

2. **insights**: dict com:
   - `melhor_regiao`: tupla (nome, receita)
   - `melhor_categoria`: tupla (nome, num_vendas)
   - `perfil_cliente_ideal`: dict com faixa_etaria e tempo_cliente mais lucrativos
   - `periodo_pico`: mês com mais vendas

3. **dashboard**: matplotlib Figure 2×3 com:
   - [0,0]: Scatter plot valor vs quantidade
   - [0,1]: Boxplot valor_total por faixa_etaria
   - [0,2]: Bar chart vendas por mês
   - [1,0]: Scatter valor_total vs idade (com cor por região)
   - [1,1]: Violin plot valor_total por região
   - [1,2]: Texto com resumo executivo

4. **recomendacoes**: lista de 5 recomendações estratégicas
   baseadas na análise

**Tempo estimado:** 10 minutos

````python
def dashboard_final_eda(df):
    """
    Dashboard final com correlações, insights e recomendações.

    Args:
        df (pd.DataFrame): DataFrame completo

    Returns:
        dict: Dashboard completo
    """
    # === SEU CÓDIGO AQUI ===

    # 1. Correlações
    corr_matrix = df.select_dtypes(include=[np.number]).corr()
    pares_fortes = []
    for i in corr_matrix.columns:
        for j in corr_matrix.columns:
            if i != j:
                corr = corr_matrix.loc[i, j]
                if abs(corr) >= 0.5:
                    pares_fortes.append((i, j, round(corr, 3)))

    fig_corr, ax = plt.subplots(figsize=(6, 4))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax)
    ax.set_title('Matriz de Correlação')
    correlacoes = {
        'matriz': corr_matrix,
        'pares_fortes': pares_fortes,
        'heatmap': fig_corr
    }

    # 2. Insights
    regiao_lucro = df.groupby('regiao')['valor_total'].sum()
    melhor_regiao = (regiao_lucro.idxmax(), regiao_lucro.max())

    cat_vendas = df.groupby('categoria')['venda_id'].count()
    melhor_categoria = (cat_vendas.idxmax(), cat_vendas.max())

    perfil = df.groupby(['faixa_etaria'])['valor_total'].mean().idxmax()
    tempo_lucr = df.groupby('tempo_cliente_meses')['valor_total'].mean().idxmax()
    perfil_cliente_ideal = {'faixa_etaria': perfil, 'tempo_cliente_meses': tempo_lucr}

    periodo_pico = df['mes'].value_counts().idxmax()

    insights = {
        'melhor_regiao': melhor_regiao,
        'melhor_categoria': melhor_categoria,
        'perfil_cliente_ideal': perfil_cliente_ideal,
        'periodo_pico': periodo_pico
    }

    # 3. Dashboard Visual
    fig, axes = plt.subplots(2, 3, figsize=(16, 8))

    sns.scatterplot(x='valor', y='quantidade', data=df, ax=axes[0, 0], color='steelblue')
    axes[0, 0].set_title('Valor vs Quantidade')

    sns.boxplot(x='faixa_etaria', y='valor_total', data=df, ax=axes[0, 1])
    axes[0, 1].set_title('Valor Total por Faixa Etária')

    vendas_mes = df.groupby('mes')['valor_total'].sum().reset_index()
    sns.barplot(x='mes', y='valor_total', data=vendas_mes, ax=axes[0, 2])
    axes[0, 2].set_title('Vendas por Mês')

    sns.scatterplot(x='idade', y='valor_total', hue='regiao', data=df, ax=axes[1, 0])
    axes[1, 0].set_title('Valor Total vs Idade por Região')

    sns.violinplot(x='regiao', y='valor_total', data=df, ax=axes[1, 1])
    axes[1, 1].set_title('Distribuição por Região')

    resumo = (
        f"Região mais lucrativa: {melhor_regiao[0]} | "
        f"Categoria destaque: {melhor_categoria[0]} | "
        f"Faixa etária ideal: {perfil_cliente_ideal['faixa_etaria']} | "
        f"Mês pico: {periodo_pico}"
    )
    axes[1, 2].text(0.1, 0.5, resumo, fontsize=11, wrap=True)
    axes[1, 2].axis('off')

    plt.tight_layout()

    # 4. Recomendações
    recomendacoes = [
        "Investir em campanhas na região mais lucrativa.",
        "Expandir o portfólio da categoria mais vendida.",
        "Criar programas de fidelização para o perfil de cliente ideal.",
        "Lançar promoções no mês de maior demanda.",
        "Monitorar variáveis fortemente correlacionadas para prever tendências."
    ]

    return {
        'correlacoes': correlacoes,
        'insights': insights,
        'dashboard': fig,
        'recomendacoes': recomendacoes
    }

    # === FIM DO SEU CÓDIGO ===
````

````python
# === TESTE Q4 ===

print("=" * 70)
print("QUESTÃO 4: DASHBOARD FINAL DE EDA")
print("=" * 70)

# Criar dashboard
dashboard = dashboard_final_eda(df_completo)

print("\n=== 1. CORRELAÇÕES ===")
print(f"\nPares com correlação forte (|r| >= 0.5):")
if dashboard['correlacoes']['pares_fortes']:
    for v1, v2, corr in dashboard['correlacoes']['pares_fortes']:
        print(f"  {v1} ↔ {v2}: {corr:+.3f}")
else:
    print("  Nenhum par com correlação >= 0.5")

print("\n=== HEATMAP DE CORRELAÇÃO ===")
dashboard['correlacoes']['heatmap']
plt.show()

print("\n=== 2. INSIGHTS PRINCIPAIS ===")
insights = dashboard['insights']
print(f"\nMelhor Região: {insights['melhor_regiao'][0]}")
print(f"  Receita: R$ {insights['melhor_regiao'][1]:,.2f}")

print(f"\nMelhor Categoria: {insights['melhor_categoria'][0]}")
print(f"  Vendas: {insights['melhor_categoria'][1]}")

print(f"\nPerfil Cliente Ideal:")
print(f"  Faixa Etária: {insights['perfil_cliente_ideal']['faixa_etaria']}")
print(f"  Tempo de Cliente: {insights['perfil_cliente_ideal']['tempo_cliente_meses']} meses")

print(f"\nPeríodo Pico: Mês {insights['periodo_pico']}")

print("\n=== 3. DASHBOARD EXECUTIVO ===")
dashboard['dashboard']
plt.show()

print("\n=== 4. RECOMENDAÇÕES ESTRATÉGICAS ===")
for i, rec in enumerate(dashboard['recomendacoes'], 1):
    print(f"{i}. {rec}")

# Verificar testes
grader.check("q4")

print("\n" + "=" * 70)
print("FIM DA QUESTÃO 4")
print("=" * 70)
````
